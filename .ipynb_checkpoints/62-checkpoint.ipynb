```python
# Перше завдання

import cv2
import numpy as np
import matplotlib.pyplot as plt

# Запитуємо користувача про шлях до зображення
image_path = input("Введіть шлях до зображення: ")

# Завантажуємо зображення
image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)

# Використовуємо фільтр Собеля для визначення границь
gradient_x = cv2.Sobel(image, cv2.CV_64F, 1, 0, ksize=3)
gradient_y = cv2.Sobel(image, cv2.CV_64F, 0, 1, ksize=3)

# Обчислюємо абсолютні значення градієнтів
gradient_x = cv2.convertScaleAbs(gradient_x)
gradient_y = cv2.convertScaleAbs(gradient_y)

# Об'єднуємо градієнти за допомогою апроксимації Гаусса
gradient = cv2.addWeighted(gradient_x, 0.5, gradient_y, 0.5, 0)

# Визначаємо кути за допомогою арктангенсу
angles = np.arctan2(gradient_y, gradient_x) * 180 / np.pi

# Візуалізуємо оригінальне зображення та результати з кутами та границями
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.imshow(image, cmap='gray')
plt.title('Оригінальне зображення')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(gradient, cmap='gray')
plt.title('Границі та кути')
plt.axis('off')

# Показуємо візуалізацію
plt.show()
```

```python
# Друге завдання

import cv2
import matplotlib.pyplot as plt
from skimage.feature import hog

# Запитуємо в користувача шлях до зображення
image_path = input("Введіть шлях до зображення: ")

# Зчитуємо зображення
image = cv2.imread(image_path)

# Конвертуємо зображення в відтінки сірого
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Витягуємо ознаки HOG
features, hog_image = hog(gray_image, visualize=True)

# Виводимо зображення та витягнуті ознаки
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 5))

ax1.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
ax1.set_title('Оригінальне зображення')
ax1.axis('off')

ax2.imshow(hog_image, cmap='gray')
ax2.set_title('HOG ознаки')
ax2.axis('off')

plt.show()
```

```python
# Третє завдання

import numpy as np
from sklearn import datasets
from skimage.feature import hog
from skimage.color import rgb2gray
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

# Завантаження датасету CIFAR-10
cifar10 = datasets.fetch_openml('CIFAR_10')

# Розділення даних на ознаки та мітки
features = np.array(cifar10.data.astype(int))
labels = np.array(cifar10.target.astype(int))

# Розбиття даних на тренувальний та тестувальний набори
X_train, X_test, y_train, y_test = train_test_split(features, labels, test_size=0.2, random_state=42)

# Функція для витягування ознак зображення за допомогою HOG
def extract_hog_features(images):
    hog_features = []
    for image in images:
        # Перетворення зображення на сірошкальне
        gray_image = rgb2gray(image.reshape((32, 32, 3)))
        # Обчислення HOG для кожного зображення
        hog_feature = hog(gray_image, block_norm='L2-Hys', pixels_per_cell=(8, 8), cells_per_block=(2, 2))
        hog_features.append(hog_feature)
    return np.array(hog_features)

# Витягування ознак зображень тренувального та тестувального наборів
X_train_hog = extract_hog_features(X_train)
X_test_hog = extract_hog_features(X_test)

# Ініціалізація та навчання класифікатора SVM
svm_classifier = SVC()
svm_classifier.fit(X_train_hog, y_train)

# Прогнозування класів для тестового набору
y_pred = svm_classifier.predict(X_test_hog)

# Обчислення метрик точності
accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred, average='macro')
recall = recall_score(y_test, y_pred, average='macro')
f1 = f1_score(y_test, y_pred, average='macro')

# Виведення результатів
print("Accuracy:", accuracy)
print("Precision:", precision)
print("Recall:", recall)
print("F1-score:", f1)
```

```python
# Четверте завдання

import cv2
from skimage.feature import hog
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
import numpy as np

import warnings

# Ігнорування попередження
warnings.filterwarnings("ignore")

# Функція для витягування ознак SIFT з зображення
def extract_sift_features(image):
    gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    sift = cv2.SIFT_create()
    keypoints, descriptors = sift.detectAndCompute(gray, None)
    return descriptors

# Функція для витягування ознак HOG з зображення
def extract_hog_features(image):
    gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    features = hog(gray, orientations=9, pixels_per_cell=(8, 8), cells_per_block=(2, 2))
    return features

# Завантаження зображення
image_path = input("Введіть шлях до зображення: ")
image = cv2.imread(image_path)

# Витягування ознак SIFT
sift_features = extract_sift_features(image)
sift_features = sift_features[:128]  # Обмеження кількості ознак SIFT до 128
sift_features = sift_features.flatten()  # Пласування ознак SIFT

# Витягування ознак HOG
hog_features = extract_hog_features(image)
hog_features = hog_features[:128]  # Обмеження кількості ознак HOG до 128

# Підготовка даних для класифікації
sift_labels = ['SIFT'] * len(sift_features)
hog_labels = ['HOG'] * len(hog_features)
labels = sift_labels + hog_labels
features = np.concatenate((sift_features, hog_features)).reshape(-1, 1)

# Класифікація та порівняння точності
classifier = KNeighborsClassifier(n_neighbors=3)
classifier.fit(features, labels)
predicted_labels = classifier.predict(features)
accuracy = accuracy_score(labels, predicted_labels)

# Розділення передбачених міток на SIFT та HOG
predicted_sift_labels = predicted_labels[:len(sift_features)]
predicted_hog_labels = predicted_labels[len(sift_features):]

print("Точність (SIFT): {:.2f}%".format(accuracy_score(sift_labels, predicted_sift_labels) * 100))
print("Точність (HOG): {:.2f}%".format(accuracy_score(hog_labels, predicted_hog_labels) * 100))

```
